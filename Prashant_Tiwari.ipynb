**Converting Dataset into pandas DataFrame**

```python
boston = pd.DataFrame(df.data, columns=df.feature_names)
boston.head()
```

```python
boston.head(20)
```

**Adding a new Column of target values to the DataFrame**

```python
boston['MEDV'] = df.target
boston.head()                          # To Print first five rows
```

```python
boston.isnull()                        # To check if any value is null or not
```

```python
boston.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

X = boston.drop('MEDV', axis =1)
Y = boston['MEDV']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.15, random_state=5)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#  Fitting model on training dataset
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
Y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train,Y_train_predict)))

print("The model Performance for training set")
print('RMSE is {}'.format(rmse))
print("\n")

# On testing set
Y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test,Y_test_predict)))

print("The model Performance for training set")
print('RMSE is {}'.format(rmse))
```
